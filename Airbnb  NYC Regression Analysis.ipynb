```python
# Feature Engineering
```

```python
df.drop(['id','host_id','name','host_name','last_review','calculated_host_listings_count','reviews_per_month'],axis=1,inplace=True)
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['room_type']=le.fit_transform(df['room_type'])
df= pd.get_dummies(df)
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
dt=pd.DataFrame(df,columns=df.columns)
dt.head()
```

```python
y=dt['price']
x=dt.drop('price',axis=1)
```

```python
x.shape,y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=50)
```

```python
# Linear Regression
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
pred=lr.predict(x_test)
```

```python
from sklearn.metrics import r2_score,mean_squared_error
```

```python
print('R-squared: ',r2_score(y_test,pred))
print('RMSE:',np.sqrt(mean_squared_error(y_test,pred)))
```

```python
# Ridge Regression
```

```python
from sklearn.linear_model import Ridge
reg=Ridge()
reg.fit(x_train,y_train)
```

```python
predreg=reg.predict(x_test)
```

```python
print('R-squared: ',r2_score(y_test,predreg))
print('RMSE :',np.sqrt(mean_squared_error(y_test,predreg)))
```

```python
# LASSO Regression
```

```python
from sklearn.linear_model import Lasso
lso=Lasso()
lso.fit(x_train,y_train)
lsopred=lso.predict(x_test)
```

```python
print('R-squared: ',r2_score(y_test,lsopred))
print('RMSE: ',np.sqrt(mean_squared_error(y_test,lsopred)))
```

```python
# ElasticNet Regression
```

```python
from sklearn.linear_model import ElasticNet
enet=ElasticNet()
enet.fit(x_train,y_train)
enetpred=enet.predict(x_test)
```

```python
print('R-squared: ',r2_score(y_test,enetpred))
print('RMSE: ',np.sqrt(mean_squared_error(y_test,enetpred)))
```
